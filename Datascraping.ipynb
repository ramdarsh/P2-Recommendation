```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&sort=user_rating,desc&start=8401'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
movie_title = []
year = []
rating = []
metascore = []
grossincome = []
votes = []
runtime = []
genre = []
certificate = []
description = []
director_stars = []

movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_title.append(name)

    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_release)

    ratings = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(ratings)

    metascores = store.find('span', class_ = 'metascore').text.replace(' ', '') if store.find('span', class_ = 'metascore') else '$$$$'
    metascore.append(metascores)

    value = store.find_all('span', attrs = {'name': 'nv'})

    vote = value[0].text
    votes.append(vote)

    gross = value[1].text if len(value) > 1 else '$$$'
    grossincome.append(gross)

    run = store.p.find('span', class_ = 'runtime').text.replace('min', '')
    runtime.append(run)

    genres = store.p.find('span', class_ = 'genre').text.replace('\n', '').replace(' ', '')
    genre.append(genres)

    cert = store.p.find('span', class_ = 'certificate').text if store.p.find('span', class_ = 'certificate') else 'none'
    certificate.append(cert)

    des = (store.find_all('p', class_ = 'text-muted'))
    desc_text = ''
    for element in des:
        desc_text = element.text.strip()
    description.append(desc_text)

    director_s = store.find('p', class_ = '')

    dir_text = director_s.get_text(strip=True)
    director_stars.append(dir_text)
```
